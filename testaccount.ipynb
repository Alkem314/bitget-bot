```python
import asyncio

async def set_leverage_for_symbol(exchange, symbol):
    print(f"Setting margin mode for {symbol}")
    await exchange.set_leverage(1, symbol, {'holdSide': 'long'})
    pprint.pprint(await exchange.set_leverage(1, symbol, {'holdSide': 'short'}))

async def main():
    tickers = await exchange.fetch_tickers()
    symbols = list(tickers.keys())
    # Nombre de requêtes par groupe
    chunk_size = 2
    # Diviser les symboles en groupes
    chunks = [symbols[i:i+chunk_size] for i in range(0, len(symbols), chunk_size)]
    for chunk in chunks:
        tasks = [set_leverage_for_symbol(exchange, symbol) for symbol in chunk]
        # Exécuter les tâches en parallèle pour chaque groupe
        await asyncio.gather(*tasks)
        # Pause de 1 seconde entre chaque groupe pour respecter la limite de l'API
        await asyncio.sleep(1)
# Exécutez la fonction principale
await main()
```

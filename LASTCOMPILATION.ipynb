```python
# CHECK EXISTING Order & SL 

async def config_striker() :

    global df_open_trade_info
    global df_closed_trade_info
    global nbr_open_trade
    global trade_nb
    global nbr_max_trade
    global trail_stop
    global alertnbr
    global alert_df

    if not df_open_trade_info.empty:

        for symbol in df_open_trade_info.index :

            response_ord = []
            response_pos = None

            params = {"stop" : True}
            response_ord.extend(await exchange.fetch_open_orders(symbol, params=params))
            await asyncio.sleep(0.005)
            response_ord.extend(await exchange.fetch_open_orders(symbol))
            await asyncio.sleep(0.01)

            response_pos = await exchange.fetch_positions([symbol])
            await asyncio.sleep(0.01)

            len_resp_ord = len(response_ord)
            len_resp_pos = len(response_pos)

            if len_resp_ord == 2 and len_resp_pos == 0 :

                if (response_ord[0]['info']['side'] == 'open_long' and response_ord[1]['info']['side'] == 'close_long'
                    ) or (response_ord[0]['info']['side'] == 'close_long' and response_ord[1]['info']['side'] == 'open_long'
                          ) or (response_ord[0]['info']['side']== 'open_short' and response_ord[1]['info']['side'] =='close_short'
                                ) or (response_ord[0]['info']['side']== 'close_short' and response_ord[1]['info']['side'] =='open_short'):
                      
                      for resp in response_ord :
                            
                            if 'id' in resp['info'] :
                            
                                if (resp['info']['side'] == 'open_long' and df_open_trade_info.loc[symbol, 'id'] != resp['info']['id']
                                    ) or (resp['info']['side'] == 'open_short' and df_open_trade_info.loc[symbol, 'id'] != resp['info']['id']) :
                                   
                                    df_open_trade_info.loc[symbol, 'id'] == resp['info']['id']
                                  
                                if (resp['info']['side'] == 'close_long' and df_open_trade_info.loc[symbol, 'sl_id'] != resp['info']['id']
                                    ) or (resp['info']['side'] == 'close_short' and df_open_trade_info.loc[symbol, 'sl_id'] != resp['info']['id']) :
                                   
                                    df_open_trade_info.loc[symbol, 'sl_id'] == resp['info']['id']

                                if (resp['info']['side'] == 'open_long' and df_open_trade_info.loc[symbol, 'id'] == None
                                    ) or (resp['info']['side'] == 'open_short' and df_open_trade_info.loc[symbol, 'id'] == None) :
                                   
                                    df_open_trade_info.loc[symbol, 'id'] == resp['info']['id']
                                  
                                if (resp['info']['side'] == 'close_long' and df_open_trade_info.loc[symbol, 'sl_id'] == None
                                    ) or (resp['info']['side'] == 'close_short' and df_open_trade_info.loc[symbol, 'sl_id'] == None) :
                                   
                                    df_open_trade_info.loc[symbol, 'sl_id'] == resp['info']['id']

            elif len_resp_ord == 1 and len_resp_pos == 1 :

                if 'id' in response_ord[0]['info'] :
                
                    if (response_ord[0]['info']['side'] == 'close_long' and df_open_trade_info.loc[symbol, 'sl_id'] != response_ord[0]['info']['id']
                        ) or (response_ord[0]['info']['side']== 'close_short' and df_open_trade_info.loc[symbol, 'sl_id'] != response_ord[0]['info']['id']) :
                    
                        df_open_trade_info.loc[symbol, 'sl_id'] == response_ord[0]['info']['id']

                    if (response_ord[0]['info']['side'] == 'close_long' and df_open_trade_info.loc[symbol, 'sl_id'] == None
                        ) or (response_ord[0]['info']['side']== 'close_short' and df_open_trade_info.loc[symbol, 'sl_id'] == None) :
                    
                        df_open_trade_info.loc[symbol, 'sl_id'] == response_ord[0]['info']['id']

            elif len_resp_ord == 1 and len_resp_pos == 0  :

                if response_ord[0]['info']['side'] == 'close_long' or response_ord[0]['info']['side'] == 'close_short' :
                    
                    response_pos = await exchange.fetch_positions([symbol])

                    await asyncio.sleep(0.005)

                    if len_resp_pos == 0 :
                        
                        try :
                        
                            params = {'stop' : True, "planType" : "normal_plan"}
                            await exchange.cancel_order(response_ord["id"], symbol, params= params)

                        except Exception as e:
                                     
                            alert = {'alert': str(e)}
                            alert_data = pd.DataFrame(alert, index=[f'alert_nbr {alertnbr} conf 1'])
                            alert_df = pd.concat([alert_df, alert_data])
                            alertnbr += 1

                        df_open_trade_info.loc[symbol, 'status'] == "bad-config"

                elif response_ord[0]['info']['side'] == 'open_long' or response_ord[0]['info']['side'] == 'open_short' :

                    df_open_trade_info.loc[symbol, 'status'] == "set_new_sl"

            elif len_resp_ord == 0 and len_resp_pos == 1  :

                df_open_trade_info.loc[symbol, 'status'] == "set_new_sl"

            elif len_resp_ord == 0 and len_resp_pos == 0  :

                df_open_trade_info.loc[symbol, 'status'] == "close"

    await asyncio.sleep(0.005)
```

```python
# CHECK TIMING Order to Position

async def time_striker() :

    global df_open_trade_info
    global df_closed_trade_info
    global nbr_open_trade
    global trade_nb
    global nbr_max_trade
    global trail_stop
    global alertnbr
    global alert_df

    if not df_open_trade_info.empty:
    
        for symbol in df_open_trade_info.index :

            response = []

            params = {"stop" : True}
            response.extend(await exchange.fetch_open_orders(symbol, params=params))
            await asyncio.sleep(0.005)
            response.extend(await exchange.fetch_open_orders(symbol))
            await asyncio.sleep(0.01)

            if len(response) == 2 :

                # pprint.pprint(f'Trying Resp bloc time')

                for resp in response :

                    if (resp['id'] and resp['info']['side'] == 'open_long' and  (datetime.timedelta(seconds=round((fulldata.index[-1] - datetime.datetime.fromtimestamp(
                        resp['timestamp']/1000)).total_seconds()))).total_seconds() > time_to_fill
                        ) or (resp['id'] and resp['info']['side'] == 'open_short' and (datetime.timedelta(seconds=round((fulldata.index[-1] - datetime.datetime.fromtimestamp(
                            resp['timestamp']/1000)).total_seconds()))).total_seconds() > time_to_fill) :

                            if df_open_trade_info["sl_id"].loc[symbol] :
                            
                                try :
                                
                                    params = {'stop' : True, "planType" : "normal_plan"}

                                    await exchange.cancel_order(df_open_trade_info["sl_id"].loc[symbol], symbol, params= params)

                                    await asyncio.sleep(0.005)

                                    await exchange.cancel_order(df_open_trade_info["id"].loc[symbol], symbol, params= params)

                                    df_open_trade_info.loc[symbol, 'status'] == "cancel"

                                    display(f'trade {symbol} close for time')

                                except Exception as e:
                                     
                                    alert = {'alert': str(e)}
                                    alert_data = pd.DataFrame(alert, index=[f'alert_nbr {alertnbr} time 1'])
                                    alert_df = pd.concat([alert_df, alert_data])
                                    alertnbr += 1

                                    display(f'trade {symbol} close for time cause Exception')

                                    df_open_trade_info.loc[symbol, 'status'] == "cancel"

                            else :

                                df_open_trade_info["sl_id"].loc[symbol] == "set_new_sl"
                                
                await asyncio.sleep(0.005)
```

```python
# NETTOYAGE df_open_trade_info

async def cleaner() :

    global df_open_trade_info
    global df_closed_trade_info
    global nbr_open_trade
    global trade_nb
    global nbr_max_trade
    global trail_stop

    if not df_open_trade_info.empty:

        for symbol in df_open_trade_info.index :

            resp_close = await exchange.fetch_canceled_and_closed_orders(symbol)

            await asyncio.sleep(0.01)

            if resp_close :

                for resp in resp_close[:5] :
        
                    if resp['orderId'] == df_open_trade_info.loc[symbol, 'id'] and resp['state'] == 'canceled' :
                        
                        df_open_trade_info.loc[symbol, 'status'] == "cancel"
        
            await asyncio.sleep(0.005)

        for symbol in df_open_trade_info.index :

            if df_open_trade_info.loc[symbol, 'status'] == "close" or df_open_trade_info.loc[symbol, 'status'] == "cancel" or df_open_trade_info.loc[symbol, 'status'] == "bad-config" :
                
                df_closed_trade_info = pd.concat([df_closed_trade_info, df_open_trade_info.loc[[symbol]]])
                
                df_open_trade_info = df_open_trade_info.drop(symbol)

                nbr_open_trade -= 1
            
                await asyncio.sleep(0.005)
```
